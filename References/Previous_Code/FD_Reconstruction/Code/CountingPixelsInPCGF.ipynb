```python
import torch
import numpy as np
import matplotlib.pyplot as plt

```

```python
Npixels = torch.load('Npixels.pt').numpy()
```

```python

bins = np.linspace(0.5,150.5,151)
bins = np.append(bins,np.inf)
plt.figure(figsize = (10,10))   
hist = plt.hist(Npixels,bins = bins,histtype = 'step',cumulative=False,density=False)
Vals = hist[0]
Bins = hist[1]

plt.xlabel('Number of pixels')
plt.ylabel('Number of Events')
plt.title('Number of pixels in PCGF Events')
plt.yscale('log')
plt.grid()

# print(Bins[1:]-0.5)
# print(Vals)
```

```python
# count the number of events with more than N

N = [50,60,70,80,90,100,110,120,130,140,150]
print(f'Total Number of Events  = {np.sum(Vals)}')
for n in N:
    print('Number of events with more than {} pixels: {}'.format(n,np.sum(Vals[n:])/np.sum(Vals)*100))
```
