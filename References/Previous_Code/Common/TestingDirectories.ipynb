```python
def list_classes_and_methods(module):
    for name, obj in inspect.getmembers(module):
        if inspect.isclass(obj):
            print(f"Class: {name}")
            for method_name, _ in inspect.getmembers(obj, predicate=inspect.isfunction):
                print(f"  - {method_name}")
```
