```python
import torch
import sys
sys.path.append('/remote/tychodata/ftairli/work/Projects/MuonSignal/Code')
import matplotlib.pyplot as plt
```

```python
variants = ['squeeze','transpose12','transpose21','permute','view','reshape']

```

```python
Losses    = {}
LossesVal = {}
for variant in variants:
    tracker = torch.load(f'Model_2_{variant}_Tracker.pt')
    Losses[variant] = tracker.EpochLoss
    LossesVal[variant] = tracker.EpochValLoss
```

```python
print(Losses)
```

```python
plt.figure(figsize=(20,20))
size = [100,80,60,40,20,10]
for variant in variants:
    plt.plot(Losses[variant],label=f'Train - {variant}',marker='o',markersize=size[variants.index(variant)])
    plt.plot(LossesVal[variant],label=f'Val   - {variant}',marker='o',markersize=size[variants.index(variant)])
plt.legend()
```
