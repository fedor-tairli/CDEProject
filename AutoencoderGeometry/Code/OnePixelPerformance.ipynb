```python
import numpy as np
import matplotlib.pyplot as plt

style_path = '/remote/tychodata/ftairli/work/MPLStyleFiles/AngularDeviation.mplstyle'
plt.style.use(style_path)
```

```python
Model_Architectures =          ['Complete','No Chi_i','No Chi_i NormTrace','NoTrace','No Chi_i No Trace']
Model_Metrics_Chi_0 = np.array([0.377     , 0.442    , 0.441              ,0.502    ,0.538             ])*180/np.pi 
Model_Metrics_Rp    = np.array([6.50      , 6.81     , 6.81               ,7.200    ,7.533             ])


fig, ax1 = plt.subplots(figsize=(10, 8))

# Plot Chi_0
ax1.bar(np.arange(len(Model_Architectures)), Model_Metrics_Chi_0, label='Chi_0', alpha=0.7, width=0.4, color='b', align='center')
ax1.set_xlabel('Model Architectures')
ax1.set_ylabel('Chi_0 Metrics', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Set x-ticks with model architecture names
ax1.set_xticks(np.arange(len(Model_Architectures)))
ax1.set_xticklabels(Model_Architectures, rotation=45, ha='right')

# Create a second y-axis for Rp
ax2 = ax1.twinx()
ax2.bar(np.arange(len(Model_Architectures)) + 0.4, Model_Metrics_Rp, label='Rp', alpha=0.7, width=0.4, color='r', align='center')
ax2.set_ylabel('Rp Metrics', color='r')
ax2.tick_params(axis='y', labelcolor='r')


# Add titles and legends
plt.title('Model Metrics Comparison')
fig.tight_layout()
plt.grid(False)
plt.show()
```
