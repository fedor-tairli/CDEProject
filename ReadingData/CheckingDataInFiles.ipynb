```python
Valid_Pixel_Directions = pd.read_csv('camera_view_table_with_HEAT_down.txt', sep=' ', header=None, index_col=None)
Valid_Pixel_Directions.columns = ['EyeID','TelID','PixelID','Theta','Phi']
Valid_Pixel_Directions = Valid_Pixel_Directions[Valid_Pixel_Directions['EyeID'] == 5]
Valid_Pixel_Directions.head()
```

```python
BackwallAngle = 273.0
AllThetas = []
AllPhis   = []
AllExpectedThetas = []
AllExpectedPhis   = []
AllPixelIDs       = []
AllTelIDs         = []
for i in range(len(PixelLevelData)):

    thisTelID   = PixelLevelData.TelID[i]
    thisPixelID = PixelLevelData.PixelID[i] -440*(thisTelID-1)
    thisTheta   = PixelLevelData.Theta[i]
    thisPhi     = PixelLevelData.Phi[i] + BackwallAngle
    

    # print('PixelID:', thisPixelID, 'TelID:', thisTelID, 'Theta:', thisTheta.round(2), 'Phi:', thisPhi.round(2),end = ' ')

    if thisPixelID in Valid_Pixel_Directions.PixelID.values:
        AllPixelIDs.append(thisPixelID+1)
        cut = (Valid_Pixel_Directions.PixelID == thisPixelID+1) & (Valid_Pixel_Directions.TelID == thisTelID)

        if cut.sum() != 1:
            # print()
            # print('Error in cut')
            continue
        ExpectedTheta = Valid_Pixel_Directions[cut].Theta.values[0]
        ExpectedPhi   = Valid_Pixel_Directions[cut].Phi.values[0]

        AllThetas.append(thisTheta)
        AllPhis.append(thisPhi)
        AllExpectedThetas.append(ExpectedTheta)
        AllExpectedPhis.append(ExpectedPhi)
        AllTelIDs.append(thisTelID)
        # print('Expected Theta:', ExpectedTheta.round(2), 'Expected Phi:', ExpectedPhi.round(2), end = ' ')
    # print()

AllThetas = np.array(AllThetas)
AllPhis   = np.array(AllPhis)
AllExpectedThetas = np.array(AllExpectedThetas)
AllExpectedPhis   = np.array(AllExpectedPhis)
AllPixelIDs = np.array(AllPixelIDs)
AllTelIDs   = np.array(AllTelIDs)

AllUniquePixelIDs = np.unique(AllPixelIDs)
# print('Unique Pixel IDs:', AllUniquePixelIDs)
# print(len(AllUniquePixelIDs))
plt.figure(figsize=(10, 10))
plt.scatter(AllExpectedThetas[AllTelIDs == 1], AllThetas[AllTelIDs == 1]-AllExpectedThetas[AllTelIDs == 1], label='Measured', alpha=0.5)
plt.scatter(AllExpectedThetas[AllTelIDs == 2], AllThetas[AllTelIDs == 2]-AllExpectedThetas[AllTelIDs == 2], label='Measured', alpha=0.5)
plt.scatter(AllExpectedThetas[AllTelIDs == 3], AllThetas[AllTelIDs == 3]-AllExpectedThetas[AllTelIDs == 3], label='Measured', alpha=0.5)
# plt.ylim(-2,2)

plt.figure(figsize=(10, 10))
AllPhiDiffs = AllPhis - AllExpectedPhis
AllPhiDiffs[AllPhiDiffs > 180] -= 360
plt.scatter(AllExpectedPhis[AllTelIDs == 1], AllPhiDiffs[AllTelIDs == 1], label='Measured', alpha=0.5)
plt.scatter(AllExpectedPhis[AllTelIDs == 2], AllPhiDiffs[AllTelIDs == 2], label='Measured', alpha=0.5)
plt.scatter(AllExpectedPhis[AllTelIDs == 3], AllPhiDiffs[AllTelIDs == 3], label='Measured', alpha=0.5)
```
